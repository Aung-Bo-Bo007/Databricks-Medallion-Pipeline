```python
import sys
import os
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, concat, lit, substring, current_timestamp

# Project path 
sys.path.append(os.path.abspath('..'))
from src.schemas import get_bronze_schema

# 
spark = SparkSession.builder.getOrCreate()
print(" Connected to Databricks Cluster")
```

```python
try:
    # Bronze
    bronze_df = spark.read.table("workspace.default.ecommerce_bronze")
    silver_df = bronze_df.withColumn("CustomerID_Masked", 
                concat(lit("****"), substring(col("CustomerID").cast("string"), -4, 4)))
    
    silver_df.write.format("delta").mode("overwrite") \
             .option("overwriteSchema", "true") \
             .saveAsTable("workspace.default.ecommerce_silver")
    
    print(" Silver Layer Processed Successfully!")
    display(silver_df.limit(5))
except Exception as e:
    print(f" Error: {e}")
```
